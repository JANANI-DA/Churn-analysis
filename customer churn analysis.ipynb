# visulization

```python
#1)CHURN DISTRIBUTION
custom_colors = {"No": "blue", "Yes": "red"}
ax = sns.countplot(x="Churn", hue="Churn", data=df, palette=custom_colors, legend=False)
plt.title("Churn Distribution")
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=2)

plt.show()
```

```python
#2)CHURN BY GENDER
custom_colors = {"No": "blue", "Yes": "red"}
ax = sns.countplot(x="gender", hue="Churn", data=df, palette=custom_colors)
plt.title("Churn by Gender")
for container in ax.containers:
    ax.bar_label(container, padding=2)
plt.show()
```

```python
#3)CHURN BY CONTRACT TYPE
custom_palette = {"No": "blue", "Yes": "red"}  
sns.countplot(x="Contract", hue="Churn", data=df, palette=custom_palette)
plt.title("Churn by Contract Type")
plt.show()
```

```python
#4)MONTHLY CHARGES VS CHURN
custom_colors = {"No": "blue", "Yes": "red"} 
sns.boxplot(x="Churn", y="MonthlyCharges", hue="Churn", data=df,palette=custom_colors)
plt.title("Monthly Charges vs Churn")
plt.show()

```

```python
#5)HEATMAP OF CORRELATION
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```
